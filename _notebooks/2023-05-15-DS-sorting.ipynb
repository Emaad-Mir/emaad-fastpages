## Practice

[75, 17, 46, 80, 67, 45, 69, 79, 40, 0]

How would you sort this list with... 

- Bubble Sort

I would compare the first and second elements of the list (75 and 17) and see which one is bigger. If the first value is greater than the second value (which in this case, it is), I would swap the two values so that 17 appears at the beginning and 75 becomes the second element of the list. I would then compare the second and third elements of the list (75 and 46). If the the second element is greater than the third (which in this case it is), I would swap the two values so that 46 becomes the second element of the list. I would repeat this process until I finally get the sorted list of numbers sorted from least to greatest. 

- Selection Sort

I would use selection sort by looking through the list and finding the "extreme" values of the list, the minimum and maximum values. In this case, the minimum value is 0 and the maximum value is 80, so I would move 0 to be at the beginning of the list and move 80 to the end of the list. I would then look through the "unsorted list" of numbers and find the minimum and maximum value of that unsorted array. In this case, the minimum value would be 17 and the maximum value would be 79, so I would move 17 so that it becomes the second lowest value in the list and would move 79 so that it becomes the second highest value of the list. I would repeat this process until we finally end up with the sorted list of numbers from least to greatest.

```python
"""
* Creator: Nighthawk Coding Society
Bubble Sort of a List with optimizations
"""

# bubble sorts a list of dictionaries, base off of provided key
def bubbleSort(list, key):
    n = len(list) - 1  # list are indexed 0 to n-1, len is n
    
    # Traverse through list with i index
    for i in range(n):
        swapped = False  # optimize code, so it exits if now swaps on inner loop

        # Inner traversal using j index
        for j in range(n-i):  # n-i as positions on right are in order in bubble
 
            # Swap if the element KeyN is greater KeyN1
            keyN = list[j].get(key)
            keyN1 = list[j+1].get(key)
            if keyN > keyN1:
                swapped = True
                list[j], list[j + 1] = list[j + 1], list[j]  # single line swap
         
        if not swapped:  # if no swaps on inner pass, list is sorted
            return  # exit function
    

if __name__ == "__main__":
    # list/dictionary sample
    list_of_people = [
    {"name": "Risa", "age": 18, "city": "New York"},
    {"name": "John", "age": 63, "city": "Eugene"},
    {"name": "Shekar", "age": 18, "city": "San Francisco"},
    {"name": "Ryan", "age": 21, "city": "Los Angeles"}
    ]
    
    # assuming uniform keys, pick 1st row as source of keys
    key_row = list_of_people[0]

    # print list as defined
    print("Original")
    print(list_of_people)
    
    for key in key_row:  # finds each key in the row
        print(key)
        bubbleSort(list_of_people, key)  # sort list of people
        print(list_of_people)
```

```python
from datetime import datetime


def selectionSort(game_list, key):
    n = len(game_list)
    
    # Traverse through the list
    for i in range(n):
        min_idx = i
        
        # Find the minimum element in the unsorted part of the list
        for j in range(i + 1, n):
            if game_list[j][key] < game_list[min_idx][key]:
                min_idx = j
            elif game_list[j][key] == game_list[min_idx][key]:
                # If the values are equal, compare the release dates
                if game_list[j]['release_date'] < game_list[min_idx]['release_date']:
                    min_idx = j
                elif game_list[j]['release_date'] == game_list[min_idx]['release_date']:
                    # If the release dates are equal, compare the prices
                    if game_list[j]['price'] < game_list[min_idx]['price']:
                        min_idx = j
        
        # Swap the found minimum element with the first element
        game_list[i], game_list[min_idx] = game_list[min_idx], game_list[i]
        
    return game_list



if __name__ == "__main__":
    # List of games (sample data)
    game_list = [
        {"name": "Call of Duty", "release_date": datetime(2003, 10, 29), "price": 49.99},
        {"name": "Mario Party 9", "release_date": datetime(2012,3,2), "price": 42.00},
        {"name": "Wii Party", "release_date": datetime(2010,7,8), "price": 39.98},
        {"name": "Mario Kart Wii", "release_date": datetime(2008,4,10), "price": 69.91},
        {"name": "Minecraft (Java)", "release_date": datetime(2011,11,18), "price": 29.99},
        {"name": "Minecraft (Bedrock)", "release_date": datetime(2011,11,18), "price": 7.00},
        {"name": "Roblox", "release_date": datetime(2006,9,1), "price": 0.00}
    ]
    
    # Print the list before sorting
    print("Original List:")
    for game in game_list:
        print(game)
    
    # Sort the game list using selection sort (default: by name)
    # Sort the game list using selection sort (default: by name)
sorted_game_list = selectionSort(game_list, 'name')

# Print the sorted list by name
print("\nSorted List by Name (Alphabetical Order):")
for game in sorted_game_list:
    print(game)

# Sort and print the list by release date (chronological order)
sorted_date_list = selectionSort(game_list, 'release_date')
print("\nSorted List by Release Date (Chronological Order):")
for game in sorted_date_list:
    print(game)

# Sort and print the list by price (ascending order)
sorted_price_list = selectionSort(game_list, 'price')
print("\nSorted List by Price (Ascending Order):")
for game in sorted_price_list:
    print(game)
```

```python
from datetime import datetime

# bubble sorts a list of dictionaries based on the provided key
def bubbleSort(lst, key):
    n = len(lst) - 1  # lists are indexed 0 to n-1, len is n
    
    # Traverse through the list with i index
    for i in range(n):
        swapped = False  # optimize code, so it exits if no swaps on inner loop

        # Inner traversal using j index
        for j in range(n-i):  # n-i as positions on the right are in order in bubble sort
 
            # Swap if the element at index j is greater than the element at index j+1
            elem_j = lst[j].get(key)
            elem_j1 = lst[j+1].get(key)
            if elem_j > elem_j1:
                swapped = True
                lst[j], lst[j+1] = lst[j+1], lst[j]  # single line swap
         
        if not swapped:  # if no swaps on the inner pass, the list is sorted
            break  # exit the loop if the list is already sorted
    
    return lst
    

if __name__ == "__main__":
    # List of games (sample data)
    game_list = [
        {"name": "Call of Duty", "release_date": datetime(2003, 10, 29), "price": 49.99},
        {"name": "Mario Party 9", "release_date": datetime(2012,3,2), "price": 42.00},
        {"name": "Wii Party", "release_date": datetime(2010,7,8), "price": 39.98},
        {"name": "Mario Kart Wii", "release_date": datetime(2008,4,10), "price": 69.91},
        {"name": "Minecraft (Java)", "release_date": datetime(2011,11,18), "price": 29.99},
        {"name": "Minecraft (Bedrock)", "release_date": datetime(2011,8,16), "price": 7.00},
        {"name": "Roblox", "release_date": datetime(2006,9,1), "price": 0.00}
    ]
    
    print("Original List:")
    for game in game_list:
        print(game)
    
    # Sort the game list using bubble sort by name in alphabetical order
    sorted_game_list = bubbleSort(game_list, 'name')
    
    # Print the sorted list by name
    print("\nSorted List by Name (Alphabetical Order):")
    for game in sorted_game_list:
        print(game)
    
    # Sort and print the list by release date (chronological order)
    sorted_date_list = bubbleSort(game_list, 'release_date')
    print("\nSorted List by Release Date (Chronological Order):")
    for game in sorted_date_list:
        print(game)
    
    # Sort and print the list by price (ascending order)
    sorted_price_list = bubbleSort(game_list, 'price')
    print("\nSorted List by Price (Ascending Order):")
    for game in sorted_price_list:
        print(game)
```
